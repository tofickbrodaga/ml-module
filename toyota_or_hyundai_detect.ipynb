```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.optim import lr_scheduler
import torchvision
from torchvision import datasets, models
from torch.utils.data import Dataset, DataLoader, Subset
import albumentations as A
from albumentations.pytorch import ToTensorV2
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tqdm.auto import tqdm
import os
from pathlib import Path
import random

data_root = Path('toyota_or_hyundai/train')
external_test_root = Path('toyota_or_hyundai/test')

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

train_tfms = A.Compose([
    A.RandomResizedCrop(size=(224, 224), scale=(0.8, 1.0)),
    A.HorizontalFlip(p=0.5),
    A.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue=0.02, p=0.6),
    A.CoarseDropout(num_holes_range=(1, 2), hole_height_range=(8, 32), hole_width_range=(8, 32), fill_value=0, p=0.5),
    A.ShiftScaleRotate(shift_limit=0.05, scale_limit=0.1, rotate_limit=20, p=0.7),
    A.Normalize(mean=(0.485, 0.456, 0.406), std=(0.229, 0.224, 0.225)),
    ToTensorV2(),
])

eval_tfms = A.Compose([
    A.Resize(height=256, width=256),
    A.CenterCrop(height=224, width=224),
    A.Normalize(mean=(0.485, 0.456, 0.406), std=(0.229, 0.224, 0.225)),
    ToTensorV2(),
])

class AlbumentationsImageFolder(Dataset):
    def __init__(self, root, transform):
        self.dataset = datasets.ImageFolder(root=root)
        self.transform = transform
        self.classes = self.dataset.classes
        self.idx_to_class = {v: k for k, v in self.dataset.class_to_idx.items()}
        
    def __len__(self):
        return len(self.dataset)
        
    def __getitem__(self, idx):
        path, label = self.dataset.samples[idx]
        image = cv2.imread(path)
        if image is None:
            print(f"Warning: Could not read image {path}. Returning black image.")
            image = np.zeros((224, 224, 3), dtype=np.uint8) 
        else:
            image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        image = self.transform(image=image)['image']
        return image, label

def get_model(device):
    model = models.resnet18(weights='IMAGENET1K_V1')
    for name, child in model.named_children():
        if name in ['layer3', 'layer4', 'fc']:
            for param in child.parameters():
                param.requires_grad = True
        else:
            for param in child.parameters():
                param.requires_grad = False
    num_ftrs = model.fc.in_features
    hidd_size = num_ftrs // 3
    model.fc = nn.Sequential(
        nn.Linear(num_ftrs, hidd_size),
        nn.ReLU(),
        nn.Linear(hidd_size, 2), # 2 класса
        nn.LogSoftmax(dim=1)
    )
    model.to(device)
    return model

def train_and_test(model, trainloader, validloader, epochs, patience=3, min_delta=0.001):
    criterion = nn.NLLLoss()
    optimizer = optim.Adam(filter(lambda p: p.requires_grad, model.parameters()), lr=0.001)
    scheduler = lr_scheduler.StepLR(optimizer, step_size=6, gamma=0.1)
    
    train_losses, val_losses = [], []
    best_val_loss = np.inf
    epochs_no_improve = 0
    
    for epoch in range(epochs):
        model.train()
        running_loss = 0.0
        train_pbar = tqdm(trainloader, desc=f"Epoch {epoch+1}/{epochs} [train]")
        
        for images, labels in train_pbar:
            images, labels = images.to(device), labels.to(device)
            optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            running_loss += loss.item()
            train_pbar.set_postfix(loss=loss.item())
        
        avg_train_loss = running_loss / len(trainloader)
        train_losses.append(avg_train_loss)

        model.eval()
        val_loss = 0.0
        with torch.no_grad():
            for images, labels in validloader:
                images, labels = images.to(device), labels.to(device)
                logps = model(images)
                loss = criterion(logps, labels)
                val_loss += loss.item()
        
        avg_val_loss = val_loss / len(validloader)
        val_losses.append(avg_val_loss)
        
        print(f"Epoch {epoch+1} Results: Train L: {avg_train_loss:.4f} | Val L: {avg_val_loss:.4f}")

        if avg_val_loss < best_val_loss - min_delta:
            best_val_loss = avg_val_loss
            epochs_no_improve = 0
        else:
            epochs_no_improve += 1
            if epochs_no_improve >= patience:
                print("Early stopping triggered.")
                break
        scheduler.step()
    return train_losses, val_losses

def test_model(model, test_data, batch_size=16):
    model.eval()
    testloader = DataLoader(test_data, batch_size=batch_size, num_workers=0)
    correct = 0
    total = 0
    with torch.no_grad():
        for images, labels in testloader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            _, preds = torch.max(outputs, 1)
            correct += (preds == labels).sum().item()
            total += labels.size(0)
    acc = correct / total
    print(f"Accuracy: {acc:.4f}")
    return acc

def make_dataloaders(data_dir, batch_size, val_split=0.15, test_split=0.15, subset_size=None):
    base_ds = AlbumentationsImageFolder(data_dir, transform=train_tfms)
    eval_ds = AlbumentationsImageFolder(data_dir, transform=eval_tfms) # Для валидации
    
    total = len(base_ds)
    current_size = total if subset_size is None else min(total, subset_size)
    
    indices = np.random.permutation(total)
    indices = indices[:current_size]

    val_n = max(1, int(current_size * val_split))
    test_n = max(1, int(current_size * test_split))
    train_n = current_size - val_n - test_n

    train_idx = indices[:train_n]
    val_idx = indices[train_n : train_n + val_n]
    test_idx = indices[train_n + val_n :]

    train_subset = Subset(base_ds, train_idx)
    val_subset = Subset(eval_ds, val_idx)
    test_subset = Subset(eval_ds, test_idx)
    
    train_loader = DataLoader(train_subset, batch_size=batch_size, shuffle=True, num_workers=0)
    val_loader = DataLoader(val_subset, batch_size=batch_size, shuffle=False, num_workers=0)
    
    return {
        'train_loader': train_loader,
        'val_loader': val_loader,
        'test_subset': test_subset,
        'sizes': {'train': len(train_subset), 'val': len(val_subset), 'test': len(test_subset)},
        'class_to_idx': base_ds.dataset.class_to_idx,
        'idx_to_class': base_ds.idx_to_class
    }

def run_experiments(experiments, epochs=6):
    results = []
    
    last_model = None 
    last_loaders = None

    if not data_root.exists():
        print(f"ОШИБКА: Папка {data_root} не найдена!")
        return pd.DataFrame(), None, None

    total_images = len(datasets.ImageFolder(data_root))
    print(f"Картинок в папке train: {total_images}")
    
    for name, subset_frac, batch_size in experiments:
        subset_size = None if subset_frac is None else max(10, int(total_images * subset_frac))
        
        print(f"\n--- Experiment: {name} (Data fraction: {subset_frac if subset_frac else 'Full'}) ---")
        
        loaders = make_dataloaders(data_root, batch_size, subset_size=subset_size)
        print(f"Samples: {loaders['sizes']}")
        
        model = get_model(device)
        train_losses, val_losses = train_and_test(model, loaders['train_loader'], loaders['val_loader'], epochs=epochs)
        
        print("Internal Test (on a part of train): ", end="")
        acc = test_model(model, loaders['test_subset'], batch_size=batch_size)
        results.append({'run': name, 'test_acc': acc})

        plt.figure(figsize=(10, 5))
        plt.plot(train_losses, label='Training Loss')
        plt.plot(val_losses, label='Validation Loss')
        plt.title(f'Loss during {name} experiment')
        plt.xlabel('Epoch')
        plt.ylabel('Loss')
        plt.legend()
        plt.grid(True)
        plt.show()

        if name == "full":
            last_model = model
            last_loaders = loaders
        
    return pd.DataFrame(results), last_model, last_loaders

def visualize_predictions(model, dataset, idx_to_class, num_images=6):
    model.eval()
    plt.figure(figsize=(12, 8))
    indices = random.sample(range(len(dataset)), min(num_images, len(dataset)))
    
    for i, idx in enumerate(indices):
        image_tensor, true_label_idx = dataset[idx]

        mean = np.array([0.485, 0.456, 0.406])
        std = np.array([0.229, 0.224, 0.225])
        
        image_np = image_tensor.cpu().numpy().transpose((1, 2, 0))
        image_np = std * image_np + mean
        image_np = np.clip(image_np, 0, 1)

        with torch.no_grad():
            output = model(image_tensor.unsqueeze(0).to(device))
            probs = torch.exp(output)
            _, predicted_label_idx = torch.max(probs, 1)
        
        true_class = idx_to_class[true_label_idx]
        predicted_class = idx_to_class[predicted_label_idx.item()]
        
        plt.subplot(int(num_images/2), 2, i + 1)
        plt.imshow(image_np)
        plt.title(f"True: {true_class}\nPred: {predicted_class}", 
                  color="green" if true_class == predicted_class else "red")
        plt.axis('off')
    
    plt.tight_layout()
    plt.show()

if __name__ == '__main__':
    batch_size_default = 16 
    experiments_config = [
        ("small", 0.25, max(4, batch_size_default // 2)), 
        ("medium", 0.50, batch_size_default),
        ("full", None, batch_size_default),
    ]

    df_results, best_model, full_loaders = run_experiments(experiments_config, epochs=12)
    
    print(df_results)
    if best_model is not None and external_test_root.exists() and full_loaders is not None:
        real_test_ds = AlbumentationsImageFolder(external_test_root, transform=eval_tfms)
        
        if len(real_test_ds) > 0:
            test_model(best_model, real_test_ds, batch_size=16)
            visualize_predictions(best_model, real_test_ds, full_loaders['idx_to_class'], num_images=8)
        else:
            print("The external test folder is empty!")
    else:
        print("\nExternal test folder not found, or model not trained, or class mapping is missing.")
```

## Загрузка обученной модели .pth и их тест на данных из интернета

```python
import torch
import torch.nn as nn
from torchvision import models, transforms
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
import os
from pathlib import Path
import math

MODEL_PATH = 'my_car_model.pth'
TEST_FOLDER = 'testing_model'
CLASSES = ['hyundai_sonata', 'toyota_land_cruiser']
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")

def load_inference_model(path):
    model = models.resnet18(weights=None)
    num_ftrs = model.fc.in_features
    hidd_size = num_ftrs // 3
    model.fc = nn.Sequential(
        nn.Linear(num_ftrs, hidd_size),
        nn.ReLU(),
        nn.Linear(hidd_size, 2),
        nn.LogSoftmax(dim=1)
    )

    if not os.path.exists(path):
        raise FileNotFoundError(f"Файл модели {path} не найден!")

    model.load_state_dict(torch.load(path, map_location=DEVICE))
    model.to(DEVICE)
    model.eval()
    return model

inference_transforms = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

def get_prediction(image_path, model):
    try:
        image = Image.open(image_path).convert('RGB')
    except Exception as e:
        print(f"Ошибка открытия файла {image_path}: {e}")
        return None, None, None

    image_tensor = inference_transforms(image).unsqueeze(0).to(DEVICE)

    with torch.no_grad():
        output = model(image_tensor)
        probabilities = torch.exp(output)
        top_p, top_class = probabilities.topk(1, dim=1)
        
    prob = top_p.item() * 100
    class_index = top_class.item()
    predicted_class = CLASSES[class_index]
    
    return image, predicted_class, prob

def visualize_folder(folder_path, model):
    folder = Path(folder_path)
    if not folder.exists():
        print(f"Папка {folder} не найдена!")
        return

    image_files = [
        f for f in folder.glob('*') 
        if f.suffix.lower() in ['.png', '.jpg', '.jpeg', '.bmp']
    ]
    
    if not image_files:
        print("В папке нет картинок!")
        return

    num_images = len(image_files)
    cols = 4
    rows = math.ceil(num_images / cols)

    plt.figure(figsize=(cols * 4, rows * 4))


    for i, img_path in enumerate(image_files):
        image, pred_class, prob = get_prediction(img_path, model)
        
        if image is None: continue
        ax = plt.subplot(rows, cols, i + 1)
        ax.imshow(image)
        ax.axis('off')

        color = "green" if prob > 80 else "orange"
        if prob < 60: color = "red"
            
        title_text = f"{pred_class}\n{prob:.1f}%"
        ax.set_title(title_text, color=color, fontsize=12, fontweight='bold')

    plt.tight_layout()
    plt.show()

try:
    loaded_model = load_inference_model(MODEL_PATH)
    visualize_folder(TEST_FOLDER, loaded_model)

except Exception as e:
    print(f"Произошла ошибка: {e}")
```
